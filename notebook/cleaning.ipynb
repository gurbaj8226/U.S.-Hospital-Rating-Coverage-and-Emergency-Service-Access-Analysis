```python
from pathlib import Path
import pandas as pd

csv_path = Path("../data") / "Hospital_General_Information (1).csv"

#Identify columns that will be kept
keep_columns = [
    "Facility ID",
    "Facility Name",
    "City/Town",
    "State",
    "Hospital Type",
    "Hospital Ownership",
    "Emergency Services",
    "Hospital overall rating",
    "Hospital overall rating footnote"
]

#read file and clean na values
df = pd.read_csv(csv_path, usecols = keep_columns, na_values = ["", "Not Available"])

#Clean column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("/", "_")
)

#Additional column flagging whether the overall rating is available, as only those that are will be analyzed
df['rating_available'] = df["hospital_overall_rating"].notna()

#Converting overall rating from String to numeric
df["hospital_overall_rating"] = pd.to_numeric(df["hospital_overall_rating"], errors = "coerce")
```

```python
#Missing values by column
df.isna().sum()

#Duplicate facility ids
df['facility_id'].duplicated().sum()

#Unique facility ids
df['facility_id'].nunique()

#Category Checks
df['hospital_type'].value_counts(dropna = False)
df['hospital_ownership'].value_counts(dropna = False)
df['emergency_services'].value_counts(dropna = False)
df['state'].value_counts(dropna = False).sort_index()
df['city_town'].value_counts(dropna = False)

#Rating and Footnote checks
df["hospital_overall_rating"].value_counts(dropna=False).sort_index()
df["hospital_overall_rating_footnote"].value_counts(dropna=False)
(df['hospital_overall_rating'].isna() & df['hospital_overall_rating_footnote'].notna()).sum()
(df['hospital_overall_rating'].notna() & df['hospital_overall_rating_footnote'].notna()).sum()

```
